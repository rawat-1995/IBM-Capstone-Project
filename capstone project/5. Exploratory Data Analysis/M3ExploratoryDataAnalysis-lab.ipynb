## Distribution


### Determine how the data is distributed


Plot the distribution curve for the column `ConvertedComp`.


The column `ConvertedComp` contains Salary converted to annual USD salaries using the exchange rate on 2019-02-01.

This assumes 12 working months and 50 working weeks.


```python
import matplotlib.pyplot as plt
import seaborn as sns
# your code goes here

#This code uses Seaborn to create a Kernel Density Estimate (KDE) plot, which is a distribution curve,
#for the 'ConvertedComp' column. The shade=True parameter fills the area under the curve.

# Create an axes-level kernel density plot
plt.figure(figsize=(10, 6))
sns.kdeplot(df['ConvertedComp'], shade=True, linewidth=3)

# Plot formatting
plt.title('Density Curve of Converted Compensation')
plt.xlabel('Annual USD Salaries')
plt.ylabel('Density')

# Show the plot
plt.show()
```

## Outliers


### Finding outliers


Find out if outliers exist in the column `ConvertedComp` using a box plot?


```python
# your code goes here

# plt.figure(figsize=(20,3))
# sns.boxplot(x=df.ConvertedComp, data=df)
# plt.show()


df['ConvertedComp'].plot(kind='box', figsize=(8, 6))

plt.title('Distribution Data of Converted Compensation')
plt.ylabel('Annual USD Salary')

plt.show()
```

Find out the Inter Quartile Range for the column `ConvertedComp`.


```python
# your code goes here
Q1 = df['ConvertedComp'].quantile(0.25)
Q3 = df['ConvertedComp'].quantile(0.75)
IQR = Q3 - Q1
print(f'Inter Quartile Range (IQR) for ConvertedComp: {IQR}')
```

Find out the upper and lower bounds.


```python
# your code goes here
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print(f'Lower Bound: {lower_bound}, Upper Bound: {upper_bound}')
```

Identify how many outliers are there in the `ConvertedComp` column.


```python
# your code goes here
outliers_count = df[(df['ConvertedComp'] < lower_bound) | (df['ConvertedComp'] > upper_bound)].shape[0]
print(f'Number of outliers in ConvertedComp: {outliers_count}')

# or


# outlier =[]
# for x in df['ConvertedComp']:
#   if ((x > upper_bound) or (x < lower_bound)):
#     outlier.append(x)
# print(' outlier in the ConvertedComp column is', len(outlier))
```

Create a new dataframe by removing the outliers from the `ConvertedComp` column.


```python
# your code goes here
# df_removed_outliers = df[(df['ConvertedComp'] >= lower_bound) & (df['ConvertedComp'] <= upper_bound)]
# df_removed_outliers
 
# or

df_removed_outlier = df['ConvertedComp'][~((df['ConvertedComp'] < lower_bound) | 
                                           (df['ConvertedComp'] > upper_bound))]

df_removed_outlier
```

## Correlation


### Finding correlation


Find the correlation between `Age` and all other numerical columns.


```python
# The original code attempts to calculate the correlation matrix using all columns in the DataFrame (including non-numeric ones).
# This might lead to a ValueError if there are non-numeric columns present.

# # Visualize the correlation matrix with a heatmap
# corr = df.corr()
# cmap = sns.color_palette("coolwarm")
# plt.figure(figsize=(8, 6))
# sns.heatmap(corr, cmap=cmap, annot=True, center=0, vmin=-1, vmax=1, square=True, linewidths=.5)
# plt.title('Correlation Matrix Heatmap')
# plt.show()

# # Print explicit correlation values for Age
# correlation_age = df.corr()['Age']
# print(f'\nCorrelation between Age and other numerical columns:\n{correlation_age}')


# Modification:
# Exclude non-numeric columns from correlation calculation to avoid potential errors.
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
corr = df[numeric_columns].corr()

# Visualize the correlation matrix with a heatmap
cmap = sns.color_palette("coolwarm")
plt.figure(figsize=(8, 6))
sns.heatmap(corr, cmap=cmap, annot=True, center=0, vmin=-1, vmax=1, square=True, linewidths=.5)
plt.title('Correlation Matrix Heatmap')
plt.show()

# Print explicit correlation values for Age using only numeric columns
correlation_age = df[numeric_columns].corr()['Age']
print(f'\nCorrelation between Age and other numerical columns:\n{correlation_age}')
```
